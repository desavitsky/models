```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/waico/SKAB/master/data/valve1/1.csv",index_col='datetime',sep=';',parse_dates=True)
```

```python
dfs = []
for i in range(0,15):
  dfs.append(pd.read_csv(f"https://raw.githubusercontent.com/waico/SKAB/master/data/valve1/{i}.csv",index_col='datetime',sep=';',parse_dates=True))

df = pd.concat(dfs)
```
